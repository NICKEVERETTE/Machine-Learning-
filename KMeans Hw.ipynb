```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
import matplotlib.pyplot as plt


X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)

# TODO determine the best k for k-means
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12))

visualizer.fit(X)
visualizer.show()
```

```python
# TODO calculate accuracy for best K
kmeans = KMeans(n_clusters = 4)
kmeans.fit(X)
clusters = kmeans.predict(X)
centers = kmeans.cluster_centers_

plt.scatter(X[:,0], X[:, 1], c='b', s=20)
plt.scatter(centers[:, 0], centers[:, 1], marker="x", s=100, linewidths=10, color="r")
```

```python
# TODO draw a confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()  # for plot styling

mat = confusion_matrix(clusters, y_true)

sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
